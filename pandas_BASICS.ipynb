# Reading Data

* iloc - gets rows (or columns) at particular positions in the index (so it only takes integers).
* loc  - gets rows (or columns) with particular labels from the index - & boolean indexing / or selection based on conditions

* eg. 
....loc....
> df.loc[df['Age'] < 30, ['Color', 'Height']]

* ....iloc...
> df.iloc[(df['Age'] < 30).values, [1, 3]]

* > ..........Condn............... , ..parameters....

```
##read headers
df.columns
```

# Sorting/Describing Data

```
df.sort_values(['Type 1','HP'],ascending=[1,0])
```

## Making Changes to Data

```
df['Total'] = df.iloc[:,4:10].sum(axis=1) 
```

```
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

#**Filtering Data**

```
new_df = df.loc[(df['Type 1'] == "Grass") & (df['Type 2'] == "Poison") & (df['HP']>70)]
new_df.reset_index(drop=True,inplace=True)
new_df.head()
```

```
#No mega Containing names 
new_df = df.loc[~(df['Name'].str.contains('Mega'))]

new_df.head()
```

```
#using regex

import re

new_df = df.loc[(df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True))]
new_df.head()
```
